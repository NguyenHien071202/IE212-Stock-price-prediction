```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Load the data
data = pd.read_csv('amzndata.csv')

# Convert 'Time' column to datetime
data['time'] = pd.to_datetime(data['time'])

fig, ax = plt.subplots(figsize=(10, 6))

def update(num):
    ax.clear()
    ax.plot(data['time'][:num], data['price'][:num], label='price')
    ax.legend()
    plt.title('Amazon Price Over Time')
    plt.xlabel('Time')
    plt.ylabel('Price')

ani = FuncAnimation(fig, update, frames=range(len(data['time'])), repeat=False)

plt.show()
```
